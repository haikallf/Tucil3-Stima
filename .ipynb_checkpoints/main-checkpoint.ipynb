```python
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.pyplot import figure

def bacaTextFIle(x):
    f = open(x,'r') 
    file = f.readlines()
    operand = []
    #melakukan replace beberapa character seperti koma, titik, dan newline
    for lines in file:
        lines = lines.replace("\n"," ")
        operand.append(lines)
    opfinal = []
    for lines in operand:
        lines = lines.split()
        opfinal.append(lines)
    return opfinal

def tetangga(x):
    matrix = []
    for i in range(2,len(x)):
        for j in range(len(x[0])):
            if(int(x[i][j]) > 0):
                matrix.append((x[0][i-2],x[0][j], x[i][j]))
    return matrix

def coordinates(x) :
    matrix =[]
    print(x[1])
    specChar = ['(', ')', ',']
    for loc in x[1]:
        loc = loc.replace('(',' ')
        loc = loc.replace(')',' ')
        loc = loc.replace(',',' ')
        matrix.append(loc)
    print("coordinates : ")
    print(matrix)
    loc = []
    for lines in matrix :
        lines = lines.split()
        loc.append(lines)
    return loc
arr = []
arr = bacaTextFIle("test.txt")
print(arr)
arrNeigh = []
arrNeigh = tetangga(arr)
print(arrNeigh)
loc = []
loc = coordinates(arr)
print(loc)
```

```python
G = nx.Graph()
for i in range (len(arr[0])):
    G.add_node(arr[0][i], pos=(int(loc[i][0]), int(loc[i][1])))
    print(pos)
    
for i in range (len(arrNeigh)):
    G.add_edge(arrNeigh[i][0], arrNeigh[i][1], weight=int(arrNeigh[i][2]))

pos=nx.get_node_attributes(G,'pos')


edgeType = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] > 0]

# nodes
nx.draw_networkx_nodes(G, pos, node_size=10000)

# edges
nx.draw_networkx_edges(G, pos, edgelist=edgeType, width=5)


# labels
nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")

plt.axis("on")
plt.show()

nx.draw(G,pos)
```
